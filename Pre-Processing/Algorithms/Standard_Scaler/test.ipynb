calculate mean and standard deviation of each column and then apply standard scaler formula for each row in every column

Drawback: try not to give time complexity of O(n^2)

```python
# calculating mean
mean = []
for col in data.columns:
    mean.append(np.mean(data[col]))
```

```python
mean
```

```python
# calculating standard deviation
std = []
for col in data.columns:
    std.append(np.sqrt(np.var(data[col])))
```

```python
std
```

```python
data
```

```python
for col in data.columns:
    print(col)
```

```python
scaled = {}
scaled_data = []
for col in data.columns:
    i = 0
    for j in range(0, len(data[col])):
        scaled_data_value = (data[col][j] - mean[i])/std[i]
        scaled_data.append(scaled_data_value)
    scaled.update({col : [scaled_data]})
    scaled_data = []
    i+=1

scaled
```

```python
scaled_dataframe = pd.DataFrame(columns=scaled.keys())
scaled_dataframe
```

```python
for key in scaled.keys():
    for val in scaled[key]:
        scaled_dataframe[key] = val[0]
```

```python
numpy_array = data.to_numpy()
numpy_array
```

```python
if np.isnan(numpy_array).all() == False:
    print("No null values")
```

```python
column_means = numpy_array.mean(axis=0)
print("Means : ",column_means)
column_std = np.sqrt(numpy_array.var(axis=0))
print("Standard_deviation : ",column_std)
```

```python
result = (numpy_array - column_means)/column_std
result
```

This resulting numpy array is a standardized version of original data having mean = 0, standard deviation = 1

```python
if standardScaler.all() == result.all():
    print(True)
else:
    print(False)
```
