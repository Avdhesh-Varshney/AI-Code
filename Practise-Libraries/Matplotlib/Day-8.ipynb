# Day 8 - Matplotlib

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

## Additional Axis Objects

```python
t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)

fig, ax1 = plt.subplots(figsize=(5.8, 3), constrained_layout=True)
p1, = ax1.plot(t, s)

ax2 = ax1.twinx()
p2, = ax2.plot(t, range(len(t)), 'C1')

ax2.legend([p1, p2], ['Cosine (left)', 'Straight (right)'])
```

## Graph on Degrees and Radians

```python
fig, ax3 = plt.subplots(figsize=(5.8, 3), constrained_layout=True)
ax3.plot(t, s)
ax3.set_xlabel('Angle [rad]')

ax4 = ax3.secondary_xaxis('top', functions=(np.rad2deg, np.deg2rad))
ax4.set_xlabel('Angle []')
```
