### 2. Add two elements to this Series:

```
key: 'BBT', value: 25.5

key: 'F51', value: 19.2
```

In response, print quotations Series to the console.

### 3. Convert the following quotations Series:

```
stocks = {

    'BBT': 25.5, 
    'F51': 19.2
}

```

to DataFrame. Reset the index and name the columns 'ticker' and 'price' respectively.

In response, print this DataFrame to the console.

````python
stocks = {
    'PLW': 387.00, 
    'CDR': 339.5, 
    'TEN': 349.5, 
    '11B': 391.0, 
    'BBT': 25.5, 
    'F51': 19.2
}
quotations = pd.Series(data=stocks)

# <---- Write your code here ---->

quotations = pd.DataFrame(quotations).reset_index()

quotations.columns = ['ticker', 'price']

print(quotations)
````

### 4. Create the following DataFrame object and assign it to the companies variable:

```
     company   price   ticker
0     Amazon  2375.0  AMZN.US
1  Microsoft   178.6  MSFT.US
2   Facebook   179.2    FB.US
```

In response, print companies DataFrame to the console.

````python
# <---- Write your code here ---->

companies = {
    'company': ['Amazon', 'Microsoft', 'Facebook'],
    'price' : [2375.0, 178.6, 179.2],
    'ticker' : ['AMZN.US', 'MSFT.US', 'FB.US']
}

companies = pd.DataFrame(companies)

print(companies)
````

---

### 5. Convert the first column of the companies DataFrame to index. In response, print companies DataFrame to the console.

````python
data_dict = {
    'company': ['Amazon', 'Microsoft', 'Facebook'],
    'price': [2375.00, 178.6, 179.2],
    'ticker': ['AMZN.US', 'MSFT.US', 'FB.US']
}

companies = pd.DataFrame(data=data_dict)

# <---- Write your code here ---->

companies = companies.set_index('company')

print(companies)
````
